```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return 2 * x + 1

x = np.linspace(-10, 10, 100) # -10 부터 10 까지 100개로 쪼개기 (-10, -9.8, -9.6, ...., 9.6, 9.8, 10) - 정의역 설정
y = f(x)                      # 치역 설정

plt.plot(x, y, label='f(x) = 2x + 1')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example of function graph')
plt.legend()      #범례 표시
plt.grid(True)    #보조선 표시
plt.show()        #보이기
```

```python
def Is_injective(f, domain):  # 앞에서 정의한 f(x) 이용,
  image = set()
  for x in domain:            # f(x) 값을 fx에 넣고 image 집합에 해당 값이 없으면 image에 추가. 만약 하나라도 같은 값이 있으면 False 출력
    fx = f(x)
    if fx in image:
      return False
    image.add(fx)
  return True

domain = range(-10, 11)       # -10에서 10까지
print("Is it injective func?", Is_injective(f, domain))
```

```python
def inverse_f(y):                     # 역함수 정의
  return (y - 1) / 2

y_values = np.linspace(-10, 10, 100)
x_values = inverse_f(y_values)

plt.plot(y_values, x_values, label = 'Inverse func f^{-1}(y) = (y - 1) / 2', color = 'red')
plt.xlabel('y')
plt.ylabel('f^{-1}(y)')
plt.title('Inverse func. graph')
plt.legend()
plt.grid(True)
plt.show()
```
